### Build the linear regression model using scikit learn in boston data to predict 'Price' based on other dependent variable.

```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
```

```python
from sklearn.datasets import load_boston
boston = load_boston()
bos = pd.DataFrame(boston.data, columns=boston['feature_names'])
bos.head()
```

```python
bos['Price'] = boston['target']
bos.head()
```

```python
bos.describe()
```

```python
bos.isnull().sum()
```

```python
bos.dtypes
```

```python
X = bos.drop(columns=["Price"])
y = bos['Price']
```

```python
from sklearn.preprocessing import StandardScaler
stdscale = StandardScaler()
x_scaler = stdscale.fit_transform(X)
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
variables = x_scaler

vif = pd.DataFrame()

vif["VIF"] = [variance_inflation_factor(variables, i) for i in range(variables.shape[1])]

vif["Features"] = X.columns

vif
```

- As we see that **"RAD"** and **"TAX"** columns have more than 5 VIF so we remove that feature.

```python
X = X.drop(columns=['RAD', 'TAX'])
X
```

```python
#from sklearn.preprocessing import StandardScaler
stdscale2 = StandardScaler()
x_scaled = stdscale2.fit_transform(X)
```

```python
x_scaled
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size=0.25, random_state=8)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
from sklearn.linear_model import LinearRegression
LinReg = LinearRegression()
LinReg.fit(x_train, y_train)
```

```python
y_pred = LinReg.predict(stdscale2.transform(x_test))
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

```python
from sklearn.linear_model import Ridge, Lasso
r = Ridge()
l =Lasso()

r.fit(x_train, y_train)
l.fit(x_train, y_train)

y_predr = r.predict(x_test)
y_predl = l.predict(x_test)
```

```python
r2_score(y_test, y_predr)
```

```python
r2_score(y_test, y_predl)
```
